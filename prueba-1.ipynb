```python
%%file ejemplo2.c

#include <stdio.h>
#include <stdlib.h>

int main()
{
    long long N;
    double *arr=(double*)malloc(N * sizeof(double));
    for ( long long i = 0; i<N;i++)
    {
        arr[i]=0;
    }
    for ( long long i = 0 ; i<N; i++)
    {
        printf("%f",arr[i]);
    }
    free(arr); ///siempre se utiliza en memoria dinamica 
}
```

```python
! gcc ejemplo2.c -o ejemplo2 && ./ejemplo2
```
